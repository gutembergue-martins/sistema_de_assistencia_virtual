```python
import webbrowser
from gtts import gTTS
from IPython.display import Audio, display, clear_output
import ipywidgets as widgets
from IPython.display import Javascript

# FunÃ§Ã£o para transformar texto em voz usando gTTS
def text_to_speech(text):
    print(f"Assistente: {text}")  # Exibe o texto no console
    tts = gTTS(text, lang='pt')  # Converte o texto em Ã¡udio (em portuguÃªs)
    tts.save("output.mp3")       # Salva o Ã¡udio em um arquivo
    display(Audio("output.mp3", autoplay=True))  # Reproduz o Ã¡udio no Colab

# FunÃ§Ã£o para abrir uma pesquisa no Wikipedia
def search_wikipedia(query):
    url = f"https://pt.wikipedia.org/wiki/{query.replace(' ', '_')}"
    webbrowser.open(url)
    text_to_speech(f"Abrindo a pÃ¡gina da Wikipedia para {query}.")

# FunÃ§Ã£o para abrir o YouTube
def open_youtube():
    webbrowser.open("https://www.youtube.com")
    text_to_speech("Abrindo o YouTube.")

# FunÃ§Ã£o para abrir o Google Maps
def open_google_maps():
    webbrowser.open("https://www.google.com/maps")
    text_to_speech("Abrindo o Google Maps.")

# FunÃ§Ã£o principal do assistente virtual
def virtual_assistant():
    text_to_speech("OlÃ¡! Como posso ajudar vocÃª hoje?")

    # Cria um widget de entrada de texto
    text_input = widgets.Text(placeholder="Digite o comando...", description="Comando:")
    send_button = widgets.Button(description="Enviar")
    output = widgets.Output()

    # FunÃ§Ã£o para processar o comando
    def on_send_button_click(b):
        with output:
            clear_output(wait=True)  # Limpa a saÃ­da anterior
            command = text_input.value.strip().lower()
            text_input.value = ""  # Limpa o campo de entrada
            print(f"VocÃª digitou: {command}")

            if not command:
                text_to_speech("NÃ£o entendi. Pode repetir?")
            elif "wiki" in command:
                query = command.replace("wiki", "").strip()
                search_wikipedia(query if query else "PÃ¡gina principal")
            elif "youtube" in command:
                open_youtube()
            elif "google maps" in command:
                open_google_maps()
            elif "sair" in command or "parar" in command:
                text_to_speech("AtÃ© logo!")
                # Fecha o assistente
                display(Javascript('google.colab.kernel.disconnect();'))
            else:
                text_to_speech("Desculpe, nÃ£o entendi o comando. Tente novamente.")

    # Associa a funÃ§Ã£o ao botÃ£o
    send_button.on_click(on_send_button_click)

    # Exibe os widgets
    display(widgets.VBox([text_input, send_button, output]))

# Executa o assistente virtual
if __name__ == "__main__":
    virtual_assistant()
```
